```python
# Download draft data

df_2007=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2007_draft.csv', header=1)
df_2008=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2008_draft.csv', header=1)
df_2009=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2009_draft.csv', header=1)
df_2010=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2010_draft.csv', header=1)
df_2011=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2011_draft.csv', header=1)
df_2012=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2012_draft.csv', header=1)
df_2013=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2013_draft.csv', header=1)
df_2014=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2014_draft.csv', header=1)
df_2015=pd.read_csv('~/Desktop/desktop-tutorial/Week_11/Draft_Data/2015_draft.csv', header=1)
```

```python
# Download biographical data

df_2007Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2007_bio.csv', header=None)
df_2008Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2008_bio.csv', header=None)
df_2009Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2009_bio.csv', header=None)
df_2010Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2010_bio.csv', header=None)
df_2011Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2011_bio.csv', header=None)
df_2012Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2012_bio.csv', header=None)
df_2013Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2013_bio.csv', header=None)
df_2014Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2014_bio.csv', header=None)
df_2015Bio=pd.read_csv('~/Desktop/desktop-tutorial/Week_11//Biographical_Data/2015_bio.csv', header=None)
```

```python
# Take a look at a single draft year dataframe

df_2007.head()
```

```python
# Create full database dataframe

drafts = [df_2007, df_2008, df_2009,
          df_2010, df_2011, df_2012,
          df_2013, df_2014, df_2015]

df_drafts = pd.concat(drafts, axis=0)
```

```python
# Drop rows beyond last pick

df_2015Bio=df_2015Bio.loc[0:210]
```

```python
# Drop extra row of forfeited draft pick from Biographical Dataframe

df_2009Bio=df_2009Bio.drop(117)
```

```python
# Create full database of biographical data

bio = [df_2007Bio, df_2008Bio, df_2009Bio,
       df_2010Bio, df_2011Bio, df_2012Bio,
       df_2013Bio, df_2014Bio, df_2015Bio]

df_bio = pd.concat(bio, axis=0)
```

```python
# Ensure both Dataframes have the same number of rows

print(df_drafts.shape)
print(df_bio.shape)
```

```python
# Reset the index

df_bio.reset_index(inplace=True)
df_drafts.reset_index(inplace=True)
```

```python
# Place Heights and Weights of draftees inside larger drafted players dataframe

df_drafts['Height'] = df_bio[7]
df_drafts['Weight'] = df_bio[8]
```

```python
# Separate League from Junior team

df_drafts[['Junior Team', 'League']] = pd.DataFrame(df_drafts['Amateur Team'].str.split('(',1).tolist(),
                                                    columns = ['Team', 'League'])
```

```python
# Drop the leftover parentheses from the League column

df_drafts['League']=df_drafts['League'].str.replace(')', '')
```

```python
# Drop the Amateur Team column now that the Junior team and league are separate

df_drafts = df_drafts.drop(['Amateur Team'], axis=1)
```

```python
# Drop superfluous index column

df_drafts = df_drafts.drop(['index'], axis=1)
```

```python
# Create draft year finished dataframes

draft_2007 = df_drafts.loc[:210]
draft_2008 = df_drafts.loc[211:421]
draft_2009 = df_drafts.loc[422:631]
draft_2010 = df_drafts.loc[632:841]
draft_2011 = df_drafts.loc[842:1052]
draft_2012 = df_drafts.loc[1053:1263]
draft_2013 = df_drafts.loc[1264:1474]
draft_2014 = df_drafts.loc[1475:1684]
draft_2015 = df_drafts.loc[1685:1895]
```
